```python
def Approx_CVP(B,t):
    ''' 
    Finding an approximation of the Closest Vector Problem, 
    Implementation of the Babai's Nearest Plane algorithm.
    Input: A matrix B and target vector t.
    Output: A vector in the Lattice generated by matrix B 
    ''' 
    n=B.nrows()
    B=B.LLL()
    b=t
    G,M=B.gram_schmidt()
    for j in range(n,0,-1):
        c=((b*G[j-1])/(G[j-1]*G[j-1])).round()
        b=b-c*B[j-1]
    return t-b
```
